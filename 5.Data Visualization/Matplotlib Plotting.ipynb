#### 2.2 Matplotlib Tutorial
# for a comprehensive list, see http://matplotlib.org/api/pyplot_summary.html

```python
%pylab inline
# This following imports are not strictly necessary: they are done by default 
# with the above pylab mode command
import numpy as np
import matplotlib.pyplot as plt
```

Histograms

```python
x = np.random.normal(size = 1000)
fig, ax = plt.subplots()
H = ax.hist(x, bins=50, alpha = 0.5, histtype ='barstacked')
```

Pie Plot

```python
fracs = [30, 15, 45, 10]
colors = ['b','g','r','w']
fig, ax = plt.subplots(figsize = (6, 6)) # make the plot square
pie = ax.pie(fracs, colors = colors, explode = (0, 0, 0.05, 0), shadow = True, 
            labels = ['A','B','C','D'])
```

```python
fracs = [30, 15, 45, 10]
colors = ['b','g','r','w']
fig, ax = plt.subplots(figsize = (6, 6)) # make the plot square
pie = ax.pie(fracs, colors = colors, explode = (0, 0, 0.1, 0), shadow = False, 
            labels = ['A','B','C','D'])
```

Errorbar Plots

```python
x = np.linspace(0, 10, 30)
dy = 0.1
y = np.random.normal(np.sin(x), dy)
fig, ax = plt.subplots()
plt.errorbar(x, y, dy, fmt='ok')
```

Filled Plots

```python
x = np.linspace(0, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, ax = plt.subplots()
ax.fill_between(x, y1, y2, where = (y1<y2), color = 'red')
ax.fill_between(x, y1, y2, where = (y1>y2), color = 'blue')
```

Scatter Plots

```python
x = np.random.random(50)
y = np.random.random(50)
c = np.random.random(50) # color of points
s = 500*np.random.random(50) # size of points

fig, ax = plt.subplots()
im = ax.scatter(x, y, c=c, s=s, cmap=plt.cm.jet)

# add a colorbar
fig.colorbar(im, ax=ax)
# set the color limits--not necessary here, but good to know how.
im.set_clim(0.0, 1.0)
```

Contour Plots

```python
x = np.linspace(0, 10, 50)
y = np.linspace(0, 20, 60)

z = np.cos(y[:, np.newaxis])*np.sin(x)
fig, ax = plt.subplots()

# filled contours
im = ax.contourf(x, y, z, 100)

# contour lines
im2 = ax.contour(x, y, z, colors='k')

fig.colorbar(im, ax = ax)
```

Showing Images

```python
I = np.random.random((100, 100))
I += np.sin(np.linspace(0, np.pi, 100))
fig, ax = plt.subplots()
im = ax.imshow(I, cmap = plt.cm.jet)
fig.colorbar(im, ax = ax)
```

2D Histograms and Hexbin

```python
x, y = np.random.normal(size = (2, 10000))

fig, ax = plt.subplots()
im = ax.hexbin(x, y, gridsize = 50)  # Make a hexagonal binning plot.
fig.colorbar(im, ax=ax)

fig, ax = plt.subplots()
H = ax.hist2d(x, y, bins = 20) # Make a 2D histogram plot.
fig.colorbar(H[3], ax = ax)
```

Polar Plots

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection = 'polar')

theta = np.linspace(0, 10*np.pi, 1000) # Return evenly spaced numbers over a specified interval.
r = np.linspace(0, 10, 1000)

ax.plot(theta, r)
```
